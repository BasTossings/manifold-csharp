```python
using ManifoldNET;
using static ManifoldNET.Manifold;
```

```python
CrossSection section = CrossSection.Square(1.0f, 1.0f, true);

Manifold.Extrude(
    crossSection: section.Polygons(), 
    height: 1f,
    slices: 0, 
    twistDegrees: 0f).DisplayGlbFile(200)
```

```python
CrossSection section = CrossSection.Circle(1f, 100);

Manifold.Extrude(
    crossSection: section.Polygons(), 
    height: 3f,
    slices: 0, 
    twistDegrees: 0f).DisplayGlbFile(200)
```
