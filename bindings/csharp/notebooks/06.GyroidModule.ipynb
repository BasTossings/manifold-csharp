```python
using ManifoldNET;

static float Gyroid(float x, float y, float z)
{
    var xi = x - Math.PI / 4.0;
    var yi = y - Math.PI / 4.0;
    var zi = z - Math.PI / 4.0;

    return 
    (float)(Math.Cos(xi) * Math.Sin(yi) +
    Math.Cos(yi) * Math.Sin(zi) +
    Math.Cos(zi) * Math.Sin(xi));
}

public static Manifold GyroidLevelSet(float level, float period, float size, int n)
{
    return Manifold.Create(
        MeshGL.LevelSet(
            Gyroid,
            new BoundingBox(new Vector3(-period, -period, -period), new Vector3(period, period, period)),
            (float)(period / n),
            (float)level,
            true
        )
        ).Scale((size / period), (size / period), (size / period));
}

public static Manifold RhombicDodecahedron(float size)
{
    double value = size * Math.Sqrt(2.0);
    var box = Manifold.Cube((float)value, (float)value, (float)value * 2f, true);
    var result = box.Rotate(90f, 45f, 0f) & box.Rotate(90f, 45f, 90f);
    return result & box.Rotate(0f, 0f, 45f);
}

public static Manifold GyroidModule(float size = 20, int n = 15)
{
    float period = (float)Math.PI * 2f;
    var result = (GyroidLevelSet(-0.4f, period, size, n) & RhombicDodecahedron(size)) - GyroidLevelSet(0.4f, period, size, n);
    return result.Rotate(-45, 0, 90).Translate(0, 0, (float)(size / Math.Sqrt(2.0)));
}

public static Manifold Run(float size = 20, int n = 15)
{
    return GyroidModule(size, n);
}
```
