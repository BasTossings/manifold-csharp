```python
using ManifoldNET;
using static ManifoldNET.Manifold;
```

# Sphereâšª

First, we cloud generate a sphere by rotate a circle.

```python
var section = CrossSection.Circle(1f, 100);
Manifold.Revolve(section.Polygons(), 100)
```

# LoopðŸ”„

When a translation apply to circle, we can generate a loop by rotate circle.

```python
var section = CrossSection
    .Circle(1f, 100)
    .Translate(10,10);
Manifold.Revolve(section.Polygons(), 100)
```

# Revolve squre

```python
var section = CrossSection
    .Square(1, 1, true)
    .Translate(10, 10);
Manifold.Revolve(section.Polygons(), 100)
```
